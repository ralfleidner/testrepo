```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "C:\\Users\\ingri\\Downloads\\cars.csv"
df = pd.read_csv(file, sep = ';')
df.head(10)
    
```

```python
df.drop(0,axis = 0, inplace=True)
df.head()
```

```python
plt.scatter(df['Weight'],df['Acceleration'])
plt.title('weight vs acceleration')
plt.xlabel('weight')
plt.ylabel('acceleration')
plt.show()
```

```python
plt.scatter(df['Horsepower'],df['Acceleration'],color = 'green')
plt.title('HP vs acceleration')
plt.xticks([50])
plt.xlabel('HP')
plt.yticks([2])
plt.ylabel('acceleration')
plt.show()
```

```python
df.describe()
```

```python
plt.scatter(df['MPG'],df['Acceleration'])
plt.title('MPG vs acceleration')
plt.xlabel('MPG')
plt.ylabel('acceleration')
plt.show()
```

```python
plt.scatter(df['MPG'],df['Displacement'])
plt.title('MPG vs Displacement')
plt.xlabel('MPG')
plt.ylabel('Displacement')
plt.show()
```
